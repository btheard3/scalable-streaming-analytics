```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the preprocessed dataset
df = pd.read_csv('../data/cleaned_events.csv')

# Display basic info and statistics
print(df.info())
print(df.describe())

# Display the first few rows
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values per Column:\n", missing_values)

# Check for empty strings (for object columns)
empty_strings = (df == "").sum()
print("Empty String Counts per Column:\n", empty_strings)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df["event"], order=df["event"].value_counts().index, palette="viridis")
plt.title("Distribution of Events")
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["timestamp"], bins=50, kde=True, color="blue")
plt.title("Activity Over Time")
plt.xlabel("Normalized Timestamp")
plt.ylabel("Frequency")
plt.show()
```
